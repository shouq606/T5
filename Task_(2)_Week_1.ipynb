### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python

plt.figure(figsize=(14, 10))

for i, column in enumerate(df.columns):

    sns.histplot(df[column], kde=True)
    plt.title(column)

#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))


city_counts = df['City'].value_counts()
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.title('Number of Traffic Jams by City')
plt.xlabel('City')
plt.ylabel('Number of Jams')
plt.xticks(rotation=90)
plt.show()
```

```python
city_jams_delay = df.groupby('City')['JamsDelay'].sum()
plt.figure(figsize=(8, 8))
plt.pie(city_jams_delay, labels=city_jams_delay.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total Jams Delay by City')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.scatterplot(x='TrafficIndexLive', y='TravelTimeLive', data=df)
plt.title('Traffic Index Live vs Travel Time Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Travel Time Live')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

corr_matrix = df[['TrafficIndexLive', 'Jamscount', 'JamsDelay', 'JamsLength',
                  'TrafficIndexWeekAgo', 'TravelTimeHistoric', 'TravelTimeLive']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
